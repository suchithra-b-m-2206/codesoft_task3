```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris=pd.read_csv('/content/sample_data/IRIS.csv')
```

```python
iris.head()
```

```python
features = iris.iloc[:, :-1]
```

```python
features
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
features_scaled
```

```python
kmeans = KMeans(n_clusters=3)
iris['cluster'] = kmeans.fit_predict(features_scaled)
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(features_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(components[:, 0], components[:, 1], c=iris['cluster'], cmap='viridis', edgecolors='k', s=80)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
user_values=[4.9,2.0,1.0,1.3]
# incase you run the notebook and want to change values uncomment the below
# user_values = [float(input("Enter Sepal length in cm: ")),
#                float(input("Enter Sepal width in cm: ")),
#                float(input("Enter Petal length in cm: ")),
#                float(input("Enter Petal width in cm: "))]
```

```python
user_values_scaled = scaler.transform([user_values])
```

```python
predicted_cluster = kmeans.predict(user_values_scaled)[0]
clusters=['Iris Setosa','Iris Versicolour','Iris Virginica']
print(f"The predicted cluster for the given values is: {clusters[predicted_cluster-1]}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
```

```python
iris=pd.read_csv('/content/sample_data/IRIS.csv')
```

```python
X = iris.iloc[:, :-1]
y = iris['species']
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_scaled, y)
```
